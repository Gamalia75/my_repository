# Задача №2

#### Для сигналізації про аварію встановлено два незалежно працюючих сигналізатори.
#### Ймовірність того, що при аварії спрацює перший сигналізатор дорівнює 0,95,
#### ця ймовірність для другого сигналізатора дорівнює 0,9.
#### Знайти ймовірність того, що при аварії спрацює:
#### а) хоча б один сигналізатор;
#### б) тільки один сигналізатор.

```python
Pa1 = 0.95 #Подія А1 - спрацює перший сигналізатор
Pa2 = 0.9  #Подія А2 - спрацює другий сигналізатор

Pb = (Pa1+Pa2) - Pa1*Pa2 #Подія В - При аварії спрацює хоча б один сигналізатор
print(f'a) При аварії спрацює хоча б один сигналізатор - вірогідність: {round(Pb*100, 2)} %')

P_a1 = 1 - Pa1 #Подія _А1 - не спрацює перший сигналізатор
P_a2 = 1 - Pa2 #Подія _А2 - не спрацює другий сигналізатор

Pc = (Pa1*P_a2) + (Pa2*P_a1) #Подія С - При аварії спрацює тільки один сигналізатор
print(f'б) При аварії спрацює тільки один сигналізатор - вірогідність: {round(Pc*100, 2)} %')

PCa1 = (Pa1*P_a2) #Подія Сa1 - При аварії спрацює тільки перший сигналізатор
print(f'в) При аварії спрацює тільки перший сигналізатор - вірогідність: {round(PCa1*100, 2)} %')

PCa2 = (Pa2*P_a1) #Подія Сa2 - При аварії спрацює тільки другий сигналізатор
print(f'г) При аварії спрацює тільки другий сигналізатор - вірогідність: {round(PCa2*100, 2)} %')
```
