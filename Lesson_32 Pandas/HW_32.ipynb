```python
import pandas as pd
```

##### 1.Створіть датафрейм розміром 15х5

```python
students_list = pd.read_excel("HW_32.xlsx")
students_list
```

##### 2.Додайте до датафрейму стовпчик, що містить дані про допуск студента на наступний курс.

```python
average_point = students_list["Cередній бал"].tolist()
pass_mark = []
for value in average_point:
    if value < 7:
        pass_mark.append(False)
    else:
        pass_mark.append(True)
students_list["Допуск"] = pass_mark
students_list
```

##### 3.Виведіть мінімальний, максимальний, середній та медіанний показники середнього балу студентів.

```python
students_list["Cередній бал"].agg(['max', 'mean', 'min', 'median'])
```

```python
students_list.groupby("Допуск")["Cередній бал"].agg(['max', 'mean', 'min', 'median'])
```

##### 4.Виведіть кількість студентів, що здали поточний екзамен з першого разу.

```python
len(students_list[students_list["Кількість спроб здачі поточного екзамену"] == 1])
```

```python
students_list[students_list["Кількість спроб здачі поточного екзамену"] == 1]
```
