# Задача №3

#### Два верстати-автомати виготовляють однакові вироби, які попадають на загальний конвеєр.
#### Продуктивність першого автомата в два раза більш ніж продуктивність другого.
#### Перший верстат-автомат виготовляє 60 % виробів вищого гатунку, а другий - 84 %.
#### а)	Знайти ймовірність того, що навмання взятий з конвеєра виріб вищого ґатунку.
#### б)	Навмання взятий з конвеєра виріб має вищий ґатунок.
####       Знайти ймовірність того, що він виготовлений на першому верстаті.

```python
# Подія А - навмання взятий з конвеєра виріб вищого ґатунку
Pa = (2/3 * 60/100) + (1/3 * 84/100)
print(f'Ймовірність того, що навмання взятий з конвеєра виріб вищого ґатунку: {round(Pa*100, 2)} %')

# Подія V1а - навмання взятий з конвеєра виріб вищого ґатунку виготовлено на першому верстаті
PV1a = (2/3 * 60/100)/Pa
print(f'Ймовірність того, що навмання взятий з конвеєра виріб вищого ґатунку i виготовлений на першому верстаті: {round(PV1a*100, 2)} %')

# Подія V2а - навмання взятий з конвеєра виріб вищого ґатунку виготовлено на другому верстаті
PV2a = (1/3 * 84/100)/Pa
print(f'Ймовірність того, що навмання взятий з конвеєра виріб вищого ґатунку i виготовлений на другому верстаті: {round(PV2a*100, 2)} %')
```
