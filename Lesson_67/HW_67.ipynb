### Створіть свій датасет на основі випадкових даних за допомогою numpy.linspace, та проведіть на ньому лінійну регресію, використовуючи інструменти tensorflow.

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
# Тепер ми можемо створити випадкові дані за допомогою np.linspace:
x = np.linspace(0, 10, 100)  # вхідні дані
y = 2 * x + 1 + np.random.randn(*x.shape)  # вихідні дані з доданням шуму
```

Ми створили масив x, що складається з 100 точок, розташованих на інтервалі від 0 до 10. Також ми створили масив y, який є лінійною функцією y = 2x + 1, з доданням нормального розподілу шуму з використанням np.random.randn.

```python
# Будуємо графік
plt.scatter(x, y)
plt.show()
```

```python
# Створюємо модель нейронної мережі
model = tf.keras.Sequential([tf.keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer=tf.keras.optimizers.Adam(1e-1), loss='mse')
```

```python
# Проводими навчання
history = model.fit(x, y, epochs=200)
```

```python
# Опис моделі 
print(model.summary())
```

```python
# Робимо прогноз
y_pred = model.predict(x)
# Виводимо графік
plt.scatter(x, y)
plt.plot(x, y_pred, color='red')
plt.show()
```
