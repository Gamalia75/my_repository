### 1.	Два економісти заповнюють документи з власними прогнозами щодо курсу певних коштовних паперів, після чого їх складають у спільну папку. Ймовірність відхилення прогнозованого значення курсу від фактичного понад допустиму похибку для першого економіста дорівнює 0,1; для другого - 0,2. Перший економіст заповнив 40 документів, другий - 60. Під час перевірки навмання взятий із папки документ виявився з суттєвою похибкою. Визначити ймовірність того, що його склав перший економіст.

#### Розв'язок:

##### Подія А - навмання взятий документ виявявся з суттєвою похибкою
##### Подія Е1 - документ заповнив перший економіст
##### Подія Е2 - документ заповнив другий економіст

#### a) Спочатку знаходимо повну ймовірність події А

##### P(E1) = 40/100 = 0,4  - документ заповнив перший економіст
##### P(E2) = 60/100 = 0,6  - документ заповнив другий економіст
##### P(А|E1) = 0,1  - допустима похибка для першого економіста 
##### P(А|E2) = 0,2  - допустима похибка для другого економіста 
##### P(А) = P(E1)*P(А|E1) + P(E2)*P(А|E2) 
##### P(А) = 0,4 * 0,1  +  0,6 * 0,2 = 0,16

#### b) За допомогою формули Байєса знаходимо ймовірнисть того, що документ з похибкою заповнив  перший економіст.

##### P(E1|A) = P(E1)*P(А|E1) / P(A)
##### P(E1|A) = (0,4 * 0,1) / 0,16 = 0,25

#### Відповідь: Ймовірнисть того, що документ з похибкою заповнив перший економіст - 25%
