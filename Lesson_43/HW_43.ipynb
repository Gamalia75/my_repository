```python
# визначаємо X та y
import pandas as pd
X = pd.DataFrame(iris['data'], columns=['sepal length', 'sepal width', 'petal length', 'petal width'])
print(X)                 
y = pd.DataFrame(iris['target'], columns=["target"])
print(y)
```

### Статистичні методи фільтрації ознак в Sklearn

#### Статистичний метод фільтрації ознак VarianceThreshold

```python
# Метод VarianceTreshold дозволяє відфільтрувати ознаки, дисперсія яких менша за задане значення.

from sklearn.feature_selection import VarianceThreshold
```

```python
# treshold= порогове_значення_дисперсії
selector = VarianceThreshold(threshold=0.4)
```

```python
selector.fit_transform(X)
```

Як бачимо, метод відібрав 3 стовбчики -  sepal length,  petal length,  petal width як найбільш найбільш значущі

#### Статистичний метод ознак фільтрації SelectKBest

```python
# Метод SelectKBest дозволяє відфільтрувати ознаки

from sklearn.feature_selection import SelectKBest, f_classif
```

```python
X = pd.DataFrame(iris['data'], columns=['sepal length', 'sepal width', 'petal length', 'petal width'])               
y = pd.DataFrame(iris['target'], columns=["target"])
```

```python
X.shape
```

```python
# на нову змінну передаємо результати роботи нашого методу SelectBest
# за критерієм f-value, k=3 - це мінімальна кількість ознак (в нашому датасеті їх 4)
# за допомогою .fit_transform(X,y) робимо попереднє навчання

X_new = SelectKBest(f_classif, k=3).fit_transform(X,y)
```

```python
# кількість ознак зменшили до 3
X_new.shape
```

```python
X_new
```

Як бачимо, метод SelectKBest також відібрав 3 стовбчики - sepal length, petal length, petal width як найбільш найбільш значущі

### Відбір ознак на основі ваг у моделі - SelectFromModel

```python
# Відбір ознак можна здійснити, попередньо оцінивши їх важливість за допомогою обраної відносно простої моделі (найчастіше лінійної або деревоподібної)

from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import LogisticRegression
```

```python
# estimator=LogisticRegression() - вказуємо, яку модель будемо використовувати
# threshold=0.2 - мінімальний рівень значущості параметрів які ми хочемо відібрати
# Дані подаються в модель для навчання методом .fit()

X = pd.DataFrame(iris['data'], columns=['sepal length', 'sepal width', 'petal length', 'petal width'])              
y = pd.DataFrame(iris['target'], columns=["target"])
sfm_selector = SelectFromModel(estimator=LogisticRegression(), threshold=0.2)
sfm_selector.fit(X, y)
```

```python
# Вивести ознаки, відібрані по параметру важливості, можна за допомогою функції .get_support
X.columns[sfm_selector.get_support()]
```

### Метод головних компонент (PCA)

```python
# PCA можна реалізувати за допомогою методу із бібліотеки sklearn.decomposition.
# При ініціалізації моделі як обов”язковий аргумент необхідно вказати кількість головних компонент, до яких ми зведемо наші дані.

from sklearn.decomposition import PCA
```

```python
# визначаємо Х

X = pd.DataFrame(iris['data'], columns=['sepal length', 'sepal width', 'petal length', 'petal width']) 
```

```python
# як обов”язковий аргумент необхідно вказати кількість головних компонент, до яких ми зведемо наші дані.

pca = PCA(n_components=3)
new_X = pca.fit_transform(X)
```

```python
new_X
```

### Сингулярне розкладання матриць (SVD)

```python
# PCA можна реалізувати за допомогою методу TruncatedSVD із бібліотеки sklearn.decomposition.
# При ініціалізації моделі як обов’язковий аргумент необхідно вказати розмірність, до якої ми хочемо звести дані

from sklearn.decomposition import TruncatedSVD
```

```python
# визначаємо Х

X = pd.DataFrame(iris['data'], columns=['sepal length', 'sepal width', 'petal length', 'petal width']) 
```

```python
svd =  TruncatedSVD(n_components = 3)
X_new = svd.fit_transform(X)
X_new
```
