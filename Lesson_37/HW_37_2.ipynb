### 2. Використовуючи датасет з файлу communities.csv, виконайте нормалізацію даних. Визначте найкращий метод нормалізації для даного набору даних.

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
train_data = pd.read_csv("communities.csv")
train_data
```

Залишаємо лише числові дані

```python
numeric_cols = [cname for cname in train_data.columns if train_data[cname].dtype in ['int64', 'float64']]
train_data = train_data[numeric_cols].copy()
train_data
```

Визначаємо Х та у

```python
X = train_data.drop("0.2.2", axis=1)
y = train_data["0.2.2"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

#### Будуємо лінійну регресію та оцінюємо якість

```python
lr = LinearRegression().fit(X_train, y_train)

print(f"Score for train set: {lr.score(X_train, y_train): .3f}")
print(f"Score for test set: {lr.score(X_test, y_test): .3f}")
```

#### L1 - Лассо 

```python
from sklearn.linear_model import Lasso
lasso = Lasso(alpha=0.5).fit(X_train, y_train)
print(f"Lasso Regression-Training set score: {lasso.score(X_train, y_train):.3f}")
print(f"Lasso Regression-Test set score: {lasso.score(X_test, y_test):.3f}")
```

```python
lasso = Lasso(alpha=0.01).fit(X_train, y_train)
print("Lasso Regression-Training set score: {:.3f}".format(lasso.score(X_train, y_train)))
print("Lasso Regression-Test set score: {:.3f}".format(lasso.score(X_test, y_test)))
```

#### L2 - Ridge

```python
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=0.1).fit(X_train, y_train)
print(f"Ridge Regression-Training set score: {ridge.score(X_train, y_train):.3f}")
print(f"Ridge Regression-Test set score: {ridge.score(X_test, y_test):.3f}")
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=1).fit(X_train, y_train)
print(f"Ridge Regression-Training set score: {ridge.score(X_train, y_train):.3f}")
print(f"Ridge Regression-Test set score: {ridge.score(X_test, y_test):.3f}")
```

#### L1 + L2 

```python
from sklearn.linear_model import ElasticNet
elastic_net = ElasticNet(alpha=0.05, l1_ratio=0.05).fit(X_train, y_train)
print(f"Elastic Net-Training set score: {elastic_net.score(X_train, y_train):.3f}")
print(f"Elastic Net-Test set score: {elastic_net.score(X_test, y_test):.3f}")
```

```python
from sklearn.linear_model import ElasticNet
elastic_net = ElasticNet(alpha=0.01, l1_ratio=0.01).fit(X_train, y_train)
print(f"Elastic Net-Training set score: {elastic_net.score(X_train, y_train):.3f}")
print(f"Elastic Net-Test set score: {elastic_net.score(X_test, y_test):.3f}")
```

#### Насправді ні один з методів нормалізації не привів до покращення показників нашої регресії. Якщо порівнювати результати роботи різних методів нормалізації, то найкращий результат в нашому прикладі показав метод L2 - Ridge.
